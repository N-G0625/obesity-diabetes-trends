```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Project Overview

This notebook analyzes trends in obesity and diabetes prevalence in the United States from 2000 to 2023. The analysis uses a simplified dataset to visualize national trends and differences between the two conditions over time. All figures are generated programmatically using Python and the notebook is designed to run from top to bottom in Google Colab.



```python
## Data Creation / Loading


years = list(range(2000, 2024))

obesity_rates = [20.0, 20.4, 20.9, 21.5, 22.1, 22.8, 23.5, 24.2, 25.0, 25.8,
26.6, 27.3, 28.0, 28.7, 29.5, 30.2, 30.9, 31.6, 32.2, 32.7,
33.1, 33.4, 33.7, 34.0]

diabetes_rates = [6.0, 6.1, 6.2, 6.3, 6.5, 6.7, 6.9, 7.1, 7.4, 7.7,
8.0, 8.3, 8.6, 8.9, 9.2, 9.5, 9.8, 10.1, 10.4, 10.6,
10.8, 11.0, 11.1, 11.2]


health_df = pd.DataFrame({
'Year': years,
'Obesity Rate (%)': obesity_rates,
'Diabetes Rate (%)': diabetes_rates
})


health_df.head()
print("Year range:", health_df['Year'].min(), "-", health_df['Year'].max())
```

```python
## Figure 1 – National Obesity and Diabetes Trends (2000–2023)



plt.figure()
plt.plot(health_df['Year'], health_df['Obesity Rate (%)'], label='Obesity Rate')
plt.plot(health_df['Year'], health_df['Diabetes Rate (%)'], label='Diabetes Rate')
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.title('Figure 1: U.S. Obesity and Diabetes Trends (2000–2023)')
plt.legend()
plt.show()
```

Explanation: This line chart illustrates changes in obesity and diabetes prevalence over time from 2000 to 2023. Both conditions show a steady upward trend, with obesity rates increasing more rapidly than diabetes rates. The similar trajectories suggest a potential relationship between the two conditions. Overall, the figure highlights the growing public health impact of obesity and diabetes in the United States.


```python
## Figure 2 – Average Prevalence Comparison

average_rates = [
health_df['Obesity Rate (%)'].mean(),
health_df['Diabetes Rate (%)'].mean()
]


labels = ['Obesity', 'Diabetes']


plt.figure()
plt.bar(labels, average_rates)
plt.ylabel('Average Percentage (%)')
plt.title('Figure 2: Average Obesity vs. Diabetes Prevalence (2000–2023)')
plt.show()
```

Explanation: This bar chart compares the average obesity and diabetes prevalence across the entire study period from 2000 to 2023. Obesity rates are significantly higher on average than diabetes rates. This difference emphasizes obesity as a widespread public health issue. However, diabetes remains a major concern due to its strong association with obesity.


```python
## Figure 3 – Yearly Difference Between Obesity and Diabetes Rates

health_df['Rate Difference (%)'] = (
health_df['Obesity Rate (%)'] - health_df['Diabetes Rate (%)']
)


plt.figure()
plt.plot(health_df['Year'], health_df['Rate Difference (%)'])
plt.xlabel('Year')
plt.ylabel('Percentage Point Difference')
plt.title('Figure 3: Difference Between Obesity and Diabetes Rates (2000–2023)')
plt.show()
```

Explanation: This chart shows how the difference between obesity and diabetes prevalence has changed over time from 2000 to 2023. The gap between the two conditions widens steadily, indicating that obesity rates have increased faster than diabetes rates. This growing difference suggests obesity may contribute to future increases in diabetes prevalence. Monitoring this gap can help guide prevention efforts.


## Summary of Findings

This analysis shows a consistent increase in both obesity and diabetes prevalence in the United States from 2000 to 2023. Obesity rates increased more rapidly than diabetes rates, leading to a widening gap between the two conditions. The parallel upward trends suggest a strong association between obesity and diabetes at the population level. These findings highlight the importance of obesity prevention in reducing future diabetes risk.
